```python
import matplotlib.pyplot as plt
import numpy as np

d = 20e-3
z = np.linspace(0, d)
epsilon = 8.854e-12
rho = 25e-3
V_0 = 22e3
V_1 = 0
sigma = 1e-4

B = V_0
A = rho/(2*epsilon)*d - B/d

V_poly = np.poly1d((-rho/(2*epsilon), A, B))
print('V(x):')
print(V_poly)
print()

E_poly = -V_poly.deriv()
print('E(x):')
print(E_poly)
print()

D_poly = epsilon * E_poly
print('D(x):')
print(D_poly)
print()

J_poly = sigma * E_poly
print('J(x):')
print(J_poly)
print()

sigma_s = -D_poly(d)
print(f'{sigma_s=:1.1e}')
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 30))

ax1.plot(z*1000, V_poly(z)/1000)
ax1.grid()
ax1.set_title('Potencial Eltrico')
ax1.set_xlabel('cota (mm)')
ax1.set_ylabel('kV')

ax2.plot(z*1000, E_poly(z)/1000000)
ax2.grid()
ax2.set_title('Campo Eltrico')
ax2.set_xlabel('cota (mm)')
ax2.set_ylabel('MV/m')

ax3.plot(z*1000, D_poly(z))
ax3.grid()
ax3.set_title('Campo de Deslocamento Eltrico')
ax3.set_xlabel('cota (mm)')
ax3.set_ylabel('')

ax4.plot(z*1000, J_poly(z)/1000)
ax4.grid()
ax4.set_title('Densidade de Corrente Eltrica')
ax4.set_xlabel('cota (mm)')
ax4.set_ylabel('kA/m^2')

plt.savefig('simulacao.eps')
plt.show()
```
