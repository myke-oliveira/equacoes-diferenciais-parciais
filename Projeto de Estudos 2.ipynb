```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook

A = 3
B = 1
MAX = 100

x = np.linspace(0, A)
y = np.linspace(0, B)
x, y = np.meshgrid(x, y)

u = sum((
    1/(A*np.sinh(n*np.pi*B/A))
    *4*A**2/(n**2*np.pi**2)*np.sin(n*np.pi/2)
    *np.sin(n*np.pi*x/A)*np.sinh(n*np.pi*y/A)
    for n in range(1, MAX)
))

fig, ax = plt.subplots(1, 1, figsize=(10,8))

pcm = ax.pcolor(x, y, u, cmap='PuBu_r')
fig.colorbar(pcm)
plt.savefig('mapadecalor.eps')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
ax=plt.gca(projection='3d')
ax.plot_surface(x, y, u)
plt.savefig('problema2-3d.eps')
plt.show()
```
