# Exercício

Desenvolva os exercícios 15, 16, 17, 18 da página 462 do livro da Biblioteca Virtual – Equações Diferenciais 8ª Ed de autoria de Nagle, Saff e Snider.

$$
\frac{\partial u}{\partial t} = \beta \frac{\partial^2 u}{\partial x^2}
$$

$$
u(x, t) = X(x) \cdot T(t)
$$

$$
X(x) \cdot T'(t) = \beta X''(x) \cdot T(t)
$$

$$
\frac{T'(t)}{T(t)} = \beta \frac{X''(x)}{X(x)}
$$

$$
\frac{T'(t)}{T(t)} = \beta \lambda
$$

$$
\frac{X''(x)}{X(x)} = \lambda
$$

$$
\begin{cases}
\frac{T'(t)}{T(t)} = \beta \lambda \\
\frac{X''(x)}{X(x)} = \lambda
\end{cases}
$$

$$
\begin{cases}
T'(t) - \beta \lambda T(t) = 0 \\
X''(x) - \lambda X(x) = 0
\end{cases}
$$

Para $ \lambda > 0 $:

$$
T(t) = C e^{\beta \lambda t}
$$

$$
X(x) = A' e^{\sqrt{\lambda} x} + B' e^{-\sqrt{\lambda} x}
$$

$$
u(x, t) = \left[
A e^{\sqrt{\lambda} x} + B e^{-\sqrt{\lambda} x}
\right]
\cdot
e^{\beta \lambda t}
$$

Para $ \lambda = 0 $:

$$
T(t) = C
$$

$$
X(x) = A' + B'x
$$

$$
u(x, t) = A + Bx
$$

Para $ \lambda < 0 $:

$$
X(t) = A' \cos \left(\sqrt{-\lambda} x\right) + B' \sin \left(\sqrt{-\lambda} x\right)
$$

$$
u(x, t) = \left[
A \cos \left(\sqrt{-\lambda} x\right) + B \sin \left(\sqrt{-\lambda} x\right)
\right]
\cdot
e^{\beta \lambda t}
$$

Caso 1 e caso 2, leva para a solução trivial.

Para o caso 3, vamos impor as condições de contorno:

$$
u(0,  t) = A e^{\beta \lambda t} = 0 \quad \rightarrow \quad A = 0
$$

$$
u(L, t) = \left[B \sin \left(\sqrt{-\lambda} L\right)
\right]
\cdot
e^{\beta \lambda t} = 0
\quad \rightarrow \sqrt{-\lambda} = \frac{n\pi}{L} \quad, n\in\mathbb{N}
$$

$$
u(x, t) = \sum_{n=0}^{\infty} \left( b_n \sin \left( \frac{n\pi}{L} x\right) \cdot e^{-\frac{n^2\beta\pi^2}{L^2} t} \right)
$$

Vamos agora impor as condições de distribuição inicial de temperatura.

**(Exemplo 15)**

$$
u(x, 0) = \sum_{n=0}^{\infty} b_n \sin \left( \frac{n\pi}{L} x\right) = 
\sin x -6 \sin {4x}
$$

$$
\begin{cases}
b_1 = 1 \\
b_4 = -6 \\
b_n = 0, n \neq 1 \wedge n \neq 4
\end{cases}
$$

$$
u(x, t) = e^{-3t} \sin x - 6 e^{-16t} \sin {4x}
$$

```python
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

u = lambda x, t: np.exp(-3*t)*np.sin(x) - 6*np.exp(-16*t)*np.sin(4*x)

def draw_u(t):
    global left, right, top, bottom
    plt.figure(2, figsize=(15,12))
    x = np.linspace(0, np.pi)
    plt.plot(x, u(x, 0), label='temperatura inicial')
    plt.plot(x, u(x, t), label='temperatura atual')
    plt.grid(which='both')
    plt.title('Temperatura')
    plt.xlabel('Cota (mm)')
    plt.ylabel('Temperatura (°C)')
    plt.legend()
    plt.show()

interactive_plot = interactive(draw_u, t=(0., .5, .01))
output = interactive_plot.children[-1]
output.layout.height = '700px'
interactive_plot
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi)
t = np.linspace(0, 0.5)
x, t = np.meshgrid(x, t)
fig = plt.figure(figsize=(15, 12))
ax=plt.axes(projection='3d')
ax.plot_surface(x, t, u(x, t))
plt.show()
```
