# Projeto de Estudos EDP

```python
import numpy as np
import matplotlib.pyplot as plt

M = 13

L = 10

x=np.linspace(0, L)
fs = [4/(n*np.pi)*np.sin(n*np.pi/2)*np.sin(1/L)*np.sin(n*np.pi/L*x) for n in range(1, M)]
```

```python

plt.figure(figsize=(10,8))
for f in fs:
    plt.plot(x, f)
plt.grid()
plt.show()
```

```python
from functools import reduce

f = reduce(lambda x, y: x + y, fs)
plt.figure(figsize=(10, 8))
plt.plot(x, f)
plt.show()
```
